# 1.Data analysis of hospital beds in INDIA

```python
# importing required libraries

# to perform numerical operations
import numpy as np

# to work with dataframes
import pandas as pd

# to visualize data
import matplotlib.pyplot as plt
```

```python
# importing data to dataframe
data = pd.read_csv("hospital_beds_india.csv")

# creating a copy of a original data
df = data.copy()

# printing data frame by calling
df
```

```python
# to check data types of columns (variables)
# we use .info() function to get the data types of each columns

df.info()



# We got 
# 4 columns of "float64" data type, 
# 2 columns of "int64" data type and 
# 6 columns of "object" data type 

```

```python
# checking for missing values
# using .isnull() function will give you the boolean data
# if the value is missing it will print True
# and if the value is present it will print False

df.isnull()


# We got the boolean values.
# True for if the value is missing
# and False if the value is not missing

# But it is difficult to scheme throungh the whole output.
```

```python
# to get the idea about how many values are missing under each variable or column
# we take the sum of missing values
# we do it by adding .sum() to existing command

df.isnull().sum()


# Here we came to know the only the county column has 
# 37 missing values

```

```python
# to get the descriptive statistics of the numerical columns
# we use .describe() function

df.describe()


# Here we got 8 staistical properties of the numerical columns 
# in the given dataframe,
# such as count, mean, standard deviation, minimum, etc

```

```python
# also to get summary of the categorical data (variables)
# we use .describe( include = "O")

df.describe(include = "O")


# Here we got the output with 4 measures
# count basically gives how many obesrvations are considered while giving you the summary
# unique represents how many unique categories are revialable under particular column
# top represents most frequently occuring category
# and frequency represents the corresponding frequency of the frequently occuring category
```

```python
# to get which value is repeating how many times under the particular column
# we use .value_counts() function for a particular column

df["year"].value_counts()



# Here we came to know that the given data is not of a particular year
# we have to analyse the data according to year
```

```python
# as we know the data is not for the same year
# we sort the data year wise
# for doing so we have to use function named .sort_values(by = "column_name")

clean_df = df.sort_values(by = "year", ignore_index = True)
clean_df


# Here we got the output and stored it in the variable called clean_df 
# Now we know that the latest data is of year 2018
# we used ignore index = True for ignoring the original index
```

```python
# it seems like that the columns named source, source_url, and county
# are not necessary in this data 
# so we drop these columns by using .drop() function

clean_df = clean_df.drop(columns = ["source", "source_url", "county", "lat", "lng"])
clean_df



# Here we got the output and stored in clean_df variable 
# we see that in the new dataframe the columns named county, source, and source_url, 
# lat, lng, are absent
#
```

```python
# to get the total information about a row using particular condition we have to use 
# the following operation

clean_df[clean_df.state == "IN"]



# as we see we got the row containing state as IN

# but if we observe the population it is 1353000000 which is the total population of india
# so we know that india does not belong to the state column 
```

```python
# as we know the clean_df yet contains data that needs to be cleaned
# so to remove the row with state IN
# we have to use the .drop() function with following operation

clean_df = clean_df.drop(index = 36, axis = 1)
clean_df


# now the dataframe seems quite ok to look at 
# now we have to visualize the data
```

```python
# to visualize data we can use matplotlib library
# since we imported this library previously we will directly use it
# to plot the graph of the state wrt beds 


plt.plot(clean_df.state, clean_df.beds, marker = ".", color = "g", label = "beds per 1000 person")
plt.title("State wise beds distribution")
plt.xlabel("State")
plt.ylabel("Beds per 1000 person")
plt.legend()
plt.show()


# here we plot a graph of beds per 1000 person wrt state 
# i.e. which state has how many beds per person 
# we used plot function of the matplotlib.pyplot library
# we mentioned x-axis variables as clean_df.state and y-axis variables as clean_df.beds
# we set marker as "." 
# using plt.title we set title for the graph
# using plt.xlabel we set x axis label of the graph
# using plt.ylabel we set y axis label of the graph
# plt.legend() is the finction for setting menu for the graph
```

```python
# to plot the graph of the state wrt population
    
plt.plot(clean_df.state, clean_df.population, label = "Population", marker = ".")
plt.title("Population distribution")
plt.xlabel("State")
plt.ylabel("Population")
plt.legend()
plt.show()





## here we plot a graph of beds per person wrt state 
# i.e. which state has how many beds per person 
# we used plot function of the matplotlib.pyplot library
# we mentioned x-axis variables as clean_df.state and y-axis variables as clean_df.population
# we set marker as "." 
# using plt.title we set title for the graph
# using plt.xlabel we set x axis label of the graph
# using plt.ylabel we set y axis label of the graph
# plt.legend() is the finction for setting menu for the graph
```

```python
# sililarly we can plot the bar graphs of the data too
# for that we use plt.bar() function

plt.bar(clean_df.state, clean_df.population, color = "orange", label = "population")
plt.title("Population distribution")
plt.xlabel("State")
plt.ylabel("Population")
plt.legend()
plt.show()
```

```python
# to figure out the year frequency we plot the histogram 
# to plot a histogram we use plt.hist() function of matplotlib library

plt.hist(clean_df.year, color = "r")





# here we figured that the most of the data is of the year 2018
```

# 2. Data analysis of hospital beds in USA

```python
# importing required libraries

# to perform numerical operations
import numpy as np

# to work with dataframes
import pandas as pd

# to visualize data
import matplotlib.pyplot as plt
```

```python
# importing data to dataframe
data1 = pd.read_csv("hospital_beds_usa.csv")

# creating a copy of a original data
df1 = data1.copy()

# printing data frame by calling
df1

```

```python
# To know the shape of the dataframe 
# the function .shape is useful

df1.shape



# And here we got got output 
# The dataframe contains 5713 rows and 12 columns
# since the data is enough large to analyse
```

```python
# to check if there is any missing value 
# use .isnull()

df1.isnull()




# as we got the output in boolean type 
# it difficult to iterate through it 
```

```python
# to know exctly how many values are missing under a particular column 
# use .isnull().sum()

df1.isnull().sum()



# here we came to know that there is no any missing value in the given dataframe
```

```python
# as we see there are columns like source and source_url which dont have any importance in this data
# so we have to remove it
# using .drop() function we will remove the unnecessary columns


clean_df1 = df1.drop(["source", "source_url", "lat", "lng"], axis = 1)
clean_df1



# as there was not any need of the columns likw source, source_url , 
# lat, lng so we removed it from the data
# and stored the cleaned data in the new variable called clean_df1
```

```python
# to get the descriptive statistics of the numerical columns of the dataframe 
# use .describe() function
# and to get the stats of the columns whose datatypes are objects
# use .describe(include = "O")

print(clean_df1.describe())
print("\n", clean_df1.describe(include = "O"))




# for numerical variables we got 8 stats 
# such as count, mean, std, min, 25%, 50%,75%, and max
# and for object type variables we got 4 stats
# such as count, unique, top, and freq
# the country column cintains only one unique value
# column named state contains 51 unique values 
# county column contains 1850  unique values
# also type and measure column contains 4 and 1 unique values respectively
```

```python
# to get the unique values of a particular column 
# we can use numpy's unique function as follows

print(np.unique(clean_df1.type))


# here we came to know that the column 'type' contains 4 values\
# it returns the list of the unique values
# similarly we can find the unique values in all the columns
```

```python
# to create the data according to the types of the beds 
# we have to create 4 different data frames

type1 = pd.DataFrame(clean_df1[clean_df1.type == "ACUTE"])
type2 = pd.DataFrame(clean_df1[clean_df1.type == "ICU"])
type3 = pd.DataFrame(clean_df1[clean_df1.type == "OTHER"])
type4 = pd.DataFrame(clean_df1[clean_df1.type == "PSYCHIATRIC"])

print(type1)
print(type2)
print(type3)
print(type4)




# here we seperated the dataframe according to the type of the beds available 
# now we have four type of data frame
# type1 contains all the ACUTE type beds data 
# type2 contains all the ICU type beds data
# type3 contains all the OTHER  type beds data
# type4 contains all PSYCHIATRIC type beds data
```

```python
#  bar graph of   state wrt ACUTE type beds

plt.bar(type1.state, type1.beds, color = "green", label = "ACUTE type beds")
plt.title("state wise ACUTE type beds distribution")
plt.xlabel("state")
plt.ylabel("ACUTE type beds")
plt.legend()
plt.show()
```

```python
# bar graph for state wrt ICU type beds 

plt.bar(type2.state, type2.beds, color =, label = "ICU type beds")
plt.title("state wise ICU type beds distribution")
plt.xlabel("state")
plt.ylabel("ICU type beds")
plt.legend()
plt.show()
```

```python
# bar graph for state wrt OTHER type beds

plt.bar(type1.state, type1.beds, color = "red", label = "OTHER type beds")
plt.title("state wise OTHER type beds distribution")
plt.xlabel("state")
plt.ylabel("OTHER type beds")
plt.legend()
plt.show()
```

```python
# bar graph for state wrt PSYCHIATRIC type beds

plt.bar(type1.state, type1.beds, color = "orange", label = "PSYCHIATRIC type beds")
plt.title("state wise PSYCHIATRIC type beds distribution")
plt.xlabel("state")
plt.ylabel("PSYCHIATRIC type beds")
plt.legend()
plt.show()
```
